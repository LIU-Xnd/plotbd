```python
import plotbd
import numpy as np
from sklearn.cluster import KMeans
```

```python
from itertools import product
coordinates = np.array(
    list(product(range(0,200), range(0,200)))
)
```

```python
kmeans = KMeans(n_clusters=500)
labels = kmeans.fit_predict(
    coordinates
)
```

```python
import seaborn as sns
sns.scatterplot(
    x=coordinates[:,0],
    y=coordinates[:,1],
    hue=labels,
    s=1,
)
```

```python
ann_points = plotbd.SpAnnPoints(
    coordinates=coordinates,
    masks=labels,
)
```

```python
ann_points
```

```python
bds = plotbd.get_boundaries(
    ann_points=ann_points,
)
```

```python
sns.scatterplot(
    x=coordinates[:,0][bds],
    y=coordinates[:,1][bds],
    s=1,
)
```
